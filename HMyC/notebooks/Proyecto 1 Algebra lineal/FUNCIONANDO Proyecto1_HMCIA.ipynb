```python
import numpy as np
import numpy.linalg as lin


# Notación matricial:

# 1.   A * x = b ----> Producto punto entre matriz de coheficientes A por vector x que  
# representa las variables, b representa coeficientes independientes

# 2.   x = inv(A) * b


print("Solución sistema de ecuaciones lineales")
ecuaciones = int(input("Introduce número de ecuaciones: "))
variables = int(input("Introduce número de variables: "))

a = []

if ecuaciones == variables: 

    for i in range (ecuaciones):
        a.append([0] * variables)

    for i in range (ecuaciones):
        for j in range (variables):
            a[i][j] = float(input("Fila {}, Columna {}: ".format(i+1 , j+1)))

    print("Matriz A =")
    for fila in a:
        print("[", end= " ")
        for elemento in fila:
            print(" {} ".format(elemento), end= " ")
        print("]")
    print()

    ##--------##

    numc = int(input("¿Cuantos coef independientes tiene la ecuación?: "))
    b =[]

    while numc <= ecuaciones:
        for k in range(numc):
            b.append(float(input("Ingrese el valor de los coef independientes: ")))
            if k == 2:
                break
        break

    print("Matriz b =" , b )

    A = np.array(a)
    B = np.array(b)

    if np.linalg.det(A) == 0: ##Para comprobar si A es invertible por determinante 
        x = None 
        print("No se puede resolver ya que |A| = 0")
    else:
        # x = (A**-1)*B
        x = np.linalg.solve(A,B)
        print("\n La matriz solución x es:", x)

# else: ##falta infinitas soluciones variables > ecuaciones





```
