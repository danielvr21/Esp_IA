## Clasificación usando Perceptrón

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
sns.set()
```

**Carga del dataset**

```python
df = pd.read_csv('/content/drive/MyDrive/2. ESPECIALIZACIÓN EN IA/CLASES UAO COLAB/Herramientas/PROYECTOS/Proyecto 3 regresión logistica/datasets/winequality-red.csv', 
                 sep=';')
df.head()
```

```python
seaborn.pairplot(df, hue = "estado")
plt.show()
```

Ya que deseamos hacer una clasificación binaria, usaremos la ultima columna del dataset `quality` para crear una nueva columna y clasificar los vinos como buenos [1] (`quality` >6.5) o malos[0] (`quality` < 6.5) 

```python
estado = []
for i in df['quality']: 
    if i > 6.5:
        i = 1
        estado.append(i)
    else:
        i = 0
        estado.append(i)
df['estado'] = estado 
```

```python
df.head(10)
```

**Dividimos el dataset para entrenamiento 80% y validación**

```python
#1599 total de filas
#Entrenamiento 80% 1279 filas

x_train = df.iloc[0:1279,0:12].values
y_train = df.iloc[0:1279,12].values
print(x_train)
print(y_train)
```

```python
#Validación 20% 320 filas

x_test = df.iloc[1279:,0:12].values
y_test = df.iloc[1279:,12].values
print(x_train)
print(y_train)
```

**Importamos Perceptron de libreria Sklearn**

```python
from sklearn.linear_model import Perceptron

p = Perceptron()
p.fit(x_train, y_train)
```

**Comprobación**

```python
# dato fila 9, clase 0
dato_test = np.array([[7.5,0.50,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.80,10.5,5]])


predic = p.predict(dato_test)
print(predic)
# 7.5	0.50	0.36	6.1	0.071	17.0	102.0	0.9978	3.35	0.80	10.5	5	
```
