```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_banknote_authentication.txt', sep=',', header=None)
df.head()
```

```python
X = df.iloc[:,0:4].values
y = df.iloc[:,4:].values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)

print(x_train[0:50,:])
print(y_train[0:50,:])

print(x_test[0:50,:])
print(y_test[0:50,:])
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
all_predict = lr.predict(x_test)
print(all_predict)

print("Datos de Validacion\n", y_test)
```
