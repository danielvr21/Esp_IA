```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('haberman.data', sep=',', header=None, names=['age','year','nodes','status'])
df.head()
```

```python
import seaborn
import matplotlib.pyplot as plt
seaborn.pairplot(df, hue="status")
plt.show()
```

```python
x_train = df.iloc[0:214, 0:3].values
y_train = df.iloc[0:214,3].values
print(x_train)

x_test = df.iloc[214:,0:3].values
y_test = df.iloc[214:,3].values
```

```python
from sklearn.linear_model import Perceptron

p = Perceptron()
p.fit(x_train, y_train)
```

```python
dato_test = np.array([[70,68,0]])

predic = p.predict(dato_test)
print(predic)
```

```python
print(x_test[0:5,:])
all_predict = p.predict(x_test)
print(all_predict)
```
